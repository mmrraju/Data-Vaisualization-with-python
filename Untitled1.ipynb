```python
'''Exploring Datasets with pandas
    Visualizing Data using Matplotlib
    Pie Charts
    Box Plots
    Scatter Plots
    Bubble Plots'''

import numpy as np
import pandas as pd

df = pd.read_excel('Canada.xlsx', sheet_name = 'Canada by Citizenship', skiprows = range (20), skipfooter = 2)
df.head()
```

```python
print('Dimension', df.shape)
```

```python
### clean up data 
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)

### let's rename the column
df.rename(columns = {'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace = True)

## make all column labels of type string
df.columns = list(map(str, df.columns))

### set the country name as index
df.set_index('Country', inplace = True)

### add 'Total' as a column
df['Total'] = df.sum(axis = 1)

years = list(map(str, range(1980, 2014)))

print('Data dimension: ', df.shape)
```

```python
################ Visualizing data using Matplotlib ##############
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

### group countries by continents and apply sum() function
df_continents = df.groupby('Continent', axis = 0).sum()

print(type(df.groupby('Continent', axis = 0)))
df_continents.head()
```

```python
### autopct create %, start angle reprsents starting point
df_continents['Total'].plot(kind = 'pie', 
                   figsize = (5, 6), 
                   autopct = '%1.1f%%', 
                   startangle = 90, 
                   shadow = True,)

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # sets the pie chart to look like circle

plt.show()
```

```python
### make clear figure
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1] # ratio for each continent with which to offset each wedge
df_continents['Total'].plot(kind = 'pie', figsize = (14, 6), 
                            autopct = '%1.1f%%', startangle = 90, shadow = True,
                           labels = None, pctdistance = 1.12, 
                            colors = color_list, explode = explode_list)
plt.title('Immigration to Canada by Continent [1980-2013]')
plt.axis('equal')

# add legend
plt.legend(labels = df_continents.index, loc = 'upper left')

plt.show()
```

```python
'''Using a pie chart, explore the proportion (percentage) of new immigrants grouped by continents in the year 2013.'''

explode_list = [0.0, 0, 0, 0.1, 0.1, 0.2]
df_continents['2013'].plot(kind = 'pie', figsize = (15, 6), autopct = '%1.1f%%', startangle = 90, shadow = True,
                          labels = None, pctdistance = 1.12, explode = explode_list)

plt.title('Immigration to Canada by Continent in 2013', y=1.12)
plt.axis('equal')

# add legend 
plt.legend(labels = df_continents.index, loc='upper left')

plt.show()
```

```python
'''Create a box plot to visualize the distribution of the top 15 countries (based on total immigration) grouped by
the decades 1980s, 1990s, and 2000s.'''
df_top15 = df.sort_values(['Total'], axis = 0, ascending = False).head()
df_top15.head()
```

```python
# create a list of all years in decades 80's, 90's and 00's
year_80s = list(map(str, range(1980, 1990)))
year_90s = list(map(str, range(1990, 2000)))
year_00s = list(map(str, range(2000, 2010)))

### slice the original dataframe to create a series for each decade

df_80s = df_top15.loc[:, year_80s].sum(axis = 1)
df_90s = df_top15.loc[:, year_90s].sum(axis = 1)
df_00s = df_top15.loc[:, year_00s].sum(axis = 1)

### merge the three series into a new data fram
new_df = pd.DataFrame({'1980s':df_80s, '1990s':df_90s, '2000s':df_00s})

new_df.head()
```

```python
new_df.plot(kind = 'box', figsize = (10, 6))
plt.title('Immigration from top 15 countries for decades 80s, 90s and 2000s')

plt.show()
```

```python
'''Note how the box plot differs from the summary table created. The box plot scans the data and identifies the outliers. In order to be an outlier, the data value must be:

larger than Q3 by at least 1.5 times the interquartile range (IQR), or,
smaller than Q1 by at least 1.5 times the IQR.
Let's look at decade 2000s as an example:

Q1 (25%) = 36,101.5
Q3 (75%) = 105,505.5
IQR = Q3 - Q1 = 69,404
Using the definition of outlier, any value that is greater than Q3 by 1.5 times IQR will be flagged as outlier.

Outlier > 105,505.5 + (1.5 * 69,404)
Outlier > 209,611.5'''

new_df = new_df.reset_index()
new_df[new_df['2000s']>209611.5]
```

```python
############# Scatter plots ##################
# We can use the sum() method to get the total population per year
df_tot = pd.DataFrame(df[years].sum(axis=0))

# change the years to type int (useful for regression later on)
df_tot.index = map(int, df_tot.index)

# reset the index to put in back in as a column in the dataframe
df_tot.reset_index(inplace = True)

# rename columns
df_tot.columns = ['year', 'total']

df_tot.head()
```

```python
df_tot.plot(kind = 'scatter', x = 'year', y ='total', figsize = (10, 6), color = 'darkblue')
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
x = df_tot['year']
y = df_tot['total']
fit = np.polyfit(x, y, deg = 1)
fit
```

```python
df_tot.plot(kind = 'scatter', x ='year', y = 'total', figsize = (10, 6), color = 'darkblue')
plt.title('Total Immigrations to Canada from 1980 - 2013')
plt.ylabel('Number of Immgrants')
plt.xlabel('Year')

### plot line of best fit
plt.plot(x, fit[0]*x + fit[1], color = 'red')
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy = (2000, 150000))

plt.show()
```

```python
'''Create a  scatter plot of the total immigration from  Denmark, Norway, and Sweden to Canada from 1980 - 2013'''
df_countries = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()

df_total = pd.DataFrame(df_countries.sum(axis = 1))
df_total.reset_index(inplace =True)

df_total.columns = ['year','total']

df_total['year'] = df_total['year'].astype(int)
df_total.head()
```
