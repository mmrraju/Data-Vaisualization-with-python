```python
'''Exploring Datasets with pandas and Prepping Data
    Visualizing Data using Matplotlib
    Area Plots
    Histograms
    Bar Charts'''
import numpy as np
import pandas as pd

df = pd.read_excel("Canada.xlsx", sheet_name = 'Canada by Citizenship', skiprows = range(20), skipfooter = 2)
df.head()
```

```python
print(df.shape)
```

```python
### Clean up the datasett to remove columns that are not informative to us for visualization
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)
df.head()
```

```python
df.rename(columns = {'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace = True)
df.head()
```

```python
### let's examine the types of column labels are same or not
all(isinstance(column, str) for column in df.columns)
```

```python
df.columns = list(map(str, df.columns))
all(isinstance(column, str) for column in df.columns)
```

```python
### set country name as index 
df.set_index('Country', inplace = True)
df.head()
```

```python
df.index.name = None
```

```python
df.head()
```

```python
df['Total'] = df.sum(axis = 1)
df.head()
```

```python
print('Data dimension: ', df.shape)
```

```python
years = list(map(str, range(1980,2014)))
years
```

```python
####### Visualizing data using Matplotlib #######

# use the inline backend to generate the plots within the browser
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

df.sort_values(['Total'], ascending = False, axis = 0, inplace = True)

# get the top 5 entries
dt_top5 = df.head(5)

# transpose the dataframe
df_top5 = dt_top5[years].transpose()
dt_top5.head()
```

```python
# let's change the index values of df_top5 to type integer for ploting
df_top5.index = df_top5.index.map(int)

df_top5.plot(kind = 'area', stacked = False, figsize = (20, 10))

plt.title("Immigration Trend of Top 5 Countries")
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
### we can make change default transparency(alpha value)
df_top5.plot(kind = 'area', alpha = .25, stacked = False , figsize = (20, 10))

plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel('Nubmer of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
### using Artist layer
ax = df_top5.plot(kind = 'area', alpha = .25, figsize = (20, 10))
ax.set_title("Immigration Trend of Top 5 Countries ")
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
'''Use the scripting layer to create a stacked area plot of the 5 countries that contributed the least to 
immigration to Canada from 1980 to 2013. Use a transparency value of 0.45.'''
# get the 5 countries with the least contribution
df_least5 = df.tail(5)

### transpose the dataframe
df_least5 = df_least5[years].transpose()
df_least5.head()

### let's change index values to type integer for ploting
df_least5.index = df_least5.index.map(int)

df_least5.plot(kind = 'area', alpha = .45, figsize = (20, 10))
plt.title('Immigration Trend of 5 Countries with least contribution to Immigration')
plt.ylabel('Number of Immigrations')
plt.xlabel('Years')
plt.show()
```

```python
'''Use the artist layer to create an unstacked area plot of the 5 countries that contributed the least to 
immigration to Canada from 1980 to 2013. Use a transparency value of 0.55.'''

ax = df_least5.plot(kind = 'area', alpha = .55, stacked = False, figsize = (20, 10))
ax.set_title('Immigration Trend of 5 Countries with least Contribution to Immigrations')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
'''What is the frequency distribution of the number (population) of new immigrants from the various countries 
to Canada in 2013?'''
### Before we proceed with creating the histogram plot, let's first examine the data split into intervals
### let's quickly view the 2013 data
df['2013'].head()
```

```python
### numpy histogram method to get the bin ranges and frequency counts as follow
count, bin_edges = np.histogram(df['2013'])
print(count)
print(bin_edges)
```

```python
df['2013'].plot(kind = 'hist', figsize = (8, 5))
plt.title('Histogram of Immigration from 194 countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
### Notice that the x-axis labels do not match with the bin size. This can be fixed by passing in a xticks keyword that contains the list of the bin sizes, as follows:

# 'bin_edges' is a list of bin intervals
count, bin_edges = np.histogram(df['2013'])
df['2013'].plot(kind = 'hist', figsize = (8,5), xticks = bin_edges )
plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
'''What is the immigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013?'''
# let's quickly view the datasets
df.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
### generate histogram
df.loc[['Denmark', 'Norway', 'Sweden'], years].plot(kind = 'hist')
```

```python
### Instead of plotting the population frequency distribution of the population for the 3 countries, pandas instead plotted the population frequency distribution for the years.

# transpose dataframe
df_t = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind = 'hist', figsize = (10, 6))
plt.title('Histograme of Immigrations for Denmark, Norway and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
### let's make some change
count, bin_edges = np.histogram(df_t, 15) # increase bin size(default )
df_t.plot(kind = 'hist', figsize = (10, 6), bins = 15, alpha = .6, xticks = bin_edges,
         color = ['coral', 'darkslateblue', 'mediumseagreen'])
plt.title('Histograme of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
'''If we do no want the plots to overlap each other, we can stack them using the stacked paramemter. 
Let's also adjust the min and max x-axis labels to remove the extra gap on the edges of the plot. 
We can pass a tuple (min,max) using the xlim paramater, as show below.'''

count, bin_edges =np.histogram(df_t, 15)
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] +10
df_t.plot(kind = 'hist', figsize = (10, 6), bins = 15, xticks = bin_edges, 
          color = ['coral', 'darkslateblue', 'mediumseagreen'], stacked = True, xlim = (xmin, xmax))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

```python
'''Use the scripting layer to display the immigration distribution for Greece, Albania, and Bulgaria for years 
1980 - 2013? Use an overlapping plot with 15 bins and a transparency value of 0.35.'''

df_cof = df.loc[['Greece', 'Albania', 'Bulgaria'], years]
df_cof = df_cof.transpose()

# let's get the x-ticks values
count, bin_edges = np.histogram(df_cof, 15)

df_cof.plot(kind = 'hist', figsize = (10, 6), bins = 15, xticks = bin_edges,
           color = ['coral', 'darkslateblue', 'mediumseagreen'])
plt.title('Histograme of Immigration from greece, Albania, and Bulgaria from 1980 - 2013')
plt.ylabel('Number of years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
'''Let's compare the number of Icelandic immigrants (country = 'Iceland') to Canada from year 1980 to 2013.'''

df_iceland = df.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind = 'bar', figsize = (10, 6), color = ['red'])
plt.title('Icelandic immigrants to Canada from 1980 - 2013')
plt.ylabel('Number of immigrants')
plt.xlabel('Year')

plt.show()
```

```python
### let's annotate on this plot using annotate method of scripting layer
df_iceland.plot(kind = 'bar', figsize = (10, 6), rot = 90) # rotate the xticks(labelled points on x-axis) by 90 degree
plt.title('Icelandic immigrants to Canada from 1980 - 2013')
plt.ylabel('Number of immigrants')
plt.xlabel('Year')

# Annotate arrow
plt.annotate('',               # s: str. Will leave it blank for no text
            xy = (32, 70),     # place head of the arrow at point (year 2012 , pop 70)
            xytext = (28, 20), # place base of the arrow at point (year 2008 , pop 20)
            xycoords = 'data', # will use the coordinate system of the object being annotated 
            arrowprops = dict(arrowstyle  = '->', connectionstyle = 'arc3', color = 'red', lw=2))

plt.show()
```

```python
'''Using the scripting layter and the df_can dataset, create a horizontal bar plot showing the total number of 
immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total 
immigrant count.'''
df.sort_values(by = 'Total', ascending = True, inplace = True)

df_top15 = df['Total'].tail(15)
df_top15
```

```python

df_top15.plot(kind = 'barh', figsize = (12, 12), color= 'steelblue')
plt.title('Top 15 Countries contributing to the Immigration to Canada between 1980 - 2013')
plt.xlabel('Number of Immigrants')

# annotate value labels to each country 
for index, value in enumerate(df_top15):
    label = format(int(value), ',')
    #  place text at the end of the bar (subtracting 47000 from x, and 0.1 from y to make fit within the bar)
    plt.annotate(label, xy = (value-47000, index - 0.1), color = 'white')
    
plt.show()
```
