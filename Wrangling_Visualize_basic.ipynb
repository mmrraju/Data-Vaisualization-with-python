```python
import numpy as np
import pandas as pd

df = pd.read_excel('Canada.xlsx', header= 0, sheet_name = 'Canada by Citizenship', skiprows = range(20), skipfooter = 2)
df.head()
```

```python
df.info(verbose = False)
df.isnull().sum()
```

```python
df.columns.tolist()
df.index.tolist()

print(type(df.columns.tolist()))
print(type(df.index.tolist()))
df.shape
```

```python
### Remove few unnecessary columns using drop() methods. pandas axis = 0 represents row(default) and axis=1 columns
df.drop(['AREA', 'REG','DEV','Type', 'Coverage'], axis = 1, inplace = True)
df.head()
```

```python
### Let's rename the column. we use rename() method by passing in a dictionary of old and new names
df.rename(columns ={'OdName' : 'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace =True)
df.columns
```

```python
### we add a column and check how many null object have in the datasets
df['Total'] = df.sum(axis = 1)
df.isnull().sum()
```

```python
### a quick summary of our dataframe using the describe() method
df.describe()
```

```python
### Pandas indexing and Selection(slicing) df['column'] return series df['column1', 'column2'] return  dataframe
df.Country
```

```python
df[['Country', 1980, 1981, 1982, 1983, 1984, 1985]]
```

```python
### to do a query by a specific things. We need to change default index(default index is 0...N) we use set_index()
### opposite of set is reset. we use df.reset_index()
df.set_index('Country', inplace = True)
```

```python
### To avoid ambuigity let's column names into strings
df.columns = list(map(str, df.columns))
[print(type(x)) for x in df.columns.values]
```

```python
### declare a variable that will allow us to easily call upon the full range
years = list(map(str, range(1980, 2014)))
years
```

```python
### create the condition boolean series
condition = df['Continent']== 'Asia'
print(condition)
```

```python
### pass this condition to the dataframe
df[condition]
```

```python
### we can pass multiple criteria in the same line
df[(df['Continent']=='Asia') & (df['Region']=='Southern Asia')]
```

```python
############ Let's Visualization tools #############
# We are using the inline backend
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
### let's plot line graph of immigration from haiti using plot()
haiti = df.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
### let's label the x and y axis and convert years to integer
haiti.index= haiti.index.map(int)
haiti.plot(kind = 'line')
plt.title("Immigration from Haiti")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")
plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

```python
### Get data from India and China
df_CI = df.loc[['India', 'China'], years]
df_CI.head()
```

```python
df_CI.plot(kind ='line')
```

```python
### above is not right so we must need to use transpose()
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.plot(kind = 'line')
plt.title("Immigration from China and India")
plt.ylabel("Number of Immigration")
plt.xlabel("Year")
plt.show()
```

```python
### top 5 countries that contributed the most to immigration to Canada
df.sort_values(by = 'Total', ascending = False, axis =0, inplace = True)

df_top5 = df.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind = 'line', figsize=(14, 8))

plt.title("Immigration Trend of Top 5 Countries")
plt.ylabel("Number of Immigrants")
plt.xlabel("Years")
```
